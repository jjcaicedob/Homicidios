```python
# A histogram is a graph showing frequency distributions.
# aio = All In One

# Import the necessary libraries 
import pandas as pd
```

```python
# Load a CSV file into a Pandas DataFrame
df = pd.read_csv("homicidios_clean.csv")
```

```python
# Return a tuple representing the dimensionality of the DataFrame.
# First value  = Number of Rows including the header.
# Second value = Number of Columns.
df.shape
```

```python
# Get the frequency table for the column CANTIDAD
df.CANTIDAD.value_counts().sort_index()
```

```python
# Using Pandas directly to get a histogram
df.plot.hist(column="CANTIDAD")
```

```python
# Using Pandas directly to get a histogram
df.plot.hist(column="CANTIDAD", by="GENERO")    
```

```python
# Using Pandas directly to get a histogram
df.plot.hist(column="CANTIDAD", by="ARMAS_MEDIOS", figsize=(8,24))    
```

```python
# Using Pandas directly to get a histogram
df.plot.hist(column="CANTIDAD", by="GRUPO_ETARIO", figsize=(8,24))    
```
